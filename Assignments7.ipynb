#### Creating a class name Product containing name , price and category. Using get_info to print those info and applying discount if required.

```python
## Making the class Product
class Product:
    def __init__(self,name,price,category):
        self.name=name
        self.price=price
        self.category=category
## Printing the product details 
    def get_info(self):
        print(f"Product: {self.name}\nPrice: {self.price}\nCategory: {self.category}")
## Appluing discount on the product and returning the discounted price
    def apply_discount(self,percent):
        self.price=self.price*percent/100
        print(f"Now the discounted price of {self.name} after applying {percent}% dicsount is {self.price}")       
```

### Task 2 :- Constructor and Encapsulation

#### Modifying the Product class and making price a private attribute and setting new prices using getter and setter method

```python
## Making the class Product
class Product:
    def __init__(self,name,price,category):
        self.name=name
        self.__price=price
        self.category=category
    
## Printing the product details 
    def get_info(self):
        print(f"Product: {self.name}\nPrice: {self.__price}\nCategory: {self.category}")
## Appluing discount on the product and returning the discounted price
    def apply_discount(self,percent):
        self.__price=self.__price*percent/100
        print(f"Now the discounted price of {self.name} after applying {percent}% dicsount is {self.__price}")      
 
    def get_price(self):  ## Getter Method
        return self.__price
    
    def set_price(self,new_price): ## Setter Method
        try:
            new_price = int(new_price)

            if new_price > 0:
                old_price = self.__price
                self.__price = new_price
                print(f"The price has been updated from {old_price} to {self.__price}")
            else:
                print("Price must be greater than 0.")

        except ValueError:
            print("Please enter a numeric value.")
```

#### The product subclass will be created along with  the additional attribute of warranty will be including and we will override the get_info to include warranty.

### Task 5 : Abstraction (using Abstract Base Class)
#### Creating abstract class payment then using two subclasses to provide process_payment

```python
from abc import ABC , abstractmethod

class Payment(ABC):
    @abstractmethod
    def process_payment(self,amount):
        pass

class CreditCardPayment(Payment):
    def process_payment(self,amount):
         print(f"The payment of {amount} has been processed via Credit Card")

class UPIPayment(Payment):
    def process_payment(self,amount):
         print(f"The payment of {amount} has been processed via UPI ")
```

### Task 7 :- Mini Project: simple Inventory System


#### Creating an inventory class, where the products object will be stored , products can be added or can be remove we can get the total value and show all products
#### Creating an store class where inventory will be store and adding new product object inside it and showing all products along with their value

```python
class Inventory():
    ## stock to store product objects
    def __init__(self):
        self.stock=[]
    ## adding product to the stock
    def add_product(self,product):
        self.stock.append(product)   
    ## removing product by name
    def remove_product(self,product):
        for prod in self.stock:
            if product==prod.name:
                self.stock.remove(prod)
                print(f"The {product} has been removed successfully")
                return
        print(f"{product} not found.")
    ## Total value of items present in the inventory
    def get_total_value(self):
        total=0
        for prod in self.stock:
            total+=prod.get_price()
        return f"The total of the inventory present is {total}"
    ## Getting all the info of the product
    def show_all_products(self):
        for prod in self.stock:
            prod.get_info()
```

```python
## Products to be placed in inventory
redmi=Product('Redmi Note 15',30000, 'Digital')
samsung = Product('Samsung Galaxy Note 26',150000,'Digital')
iphone = Product('Iphone 20',100000,'Digital')
hp = Product('HP Victus',58000 , 'Digital')
apple = Product('Macbook Air',70000,'Digital')
asus = Product("Asus Rog", 80000, 'Digital')
```

```python
## Creating the inventory object
inventory=Inventory()
```

```python
## Adding product
inventory.add_product(redmi)
inventory.add_product(samsung)
inventory.add_product(iphone)
inventory.add_product(hp)
inventory.add_product(apple)
inventory.add_product(asus)
```

```python
## Getting total values of the product
inventory.get_total_value()
```

```python
## Showing info of the product
inventory.show_all_products()
```

```python
## Removing the product by name
inventory.remove_product('Redmi Note 15')
```

```python
## Showing all info after removing
inventory.show_all_products()
```

```python
## Store name is required 
class Store:
    def __init__(self,name):
        self.name=name
        self.inventory=Inventory()
 ## For adding new product   
    def add_new_product(self,name,price,category):
        product=Product(name,price,category)
        self.inventory.add_product(product)
## SHowcasing the summary that how much item and worth does the store has
    def show_summary(self):
        print(f"The name of the store is {self.name}\n")
        print(f"The total number of items listed are {len(self.inventory.stock)}\n")
        print(f"The item worth is of total {self.inventory.get_total_value()}")
```
